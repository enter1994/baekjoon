```python
a, b = map(int, input().split())

a = str(a)
b = str(b)
result_a = 0
result_b = 0
```

```python
for i in range(len(a)):
    if i == 0:
        result_a += int(a[i])
    
    elif i == 1:
        result_a += int(a[i]) * 10
    
    elif i == 2:
        result_a += int(a[i]) * 100
print(result_a)
```

```python
for i in range(len(b)):
    if i == 0:
        result_b += int(b[i])
    
    elif i == 1:
        result_b += int(b[i]) * 10
    
    elif i == 2:
        result_b += int(b[i]) * 100
print(result_b)
```

```python
max = max(result_a, result_b)
print(max)
```
