```python
word = input().lower()
alphabet = []
for i in range(26):
    alphabet += chr(i+97)
print(word)
print(alphabet)
```

```python
num_count = []

for i in range(26):
    t = alphabet[i]
    num_count += [word.count(t)]

print(num_count)
```

```python
big = max(num_count)

if num_count.count(big) > 1:
    print('?')

else:
    num_index = num_count.index(big)
    print(alphabet[num_index].upper())
```
