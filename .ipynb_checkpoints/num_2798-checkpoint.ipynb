```python
n, m = map(int, input().split())
card = map(int, input().split())
card = list(card)
card_num = len(card)
result = []
```

```python
for i in range(card_num - 2):
    for j in range(i+1, card_num - 1):
        for k in range(j+1, card_num):
            if card[i] + card[j] + card[k] > m:
                continue
            else:
                result += [card[i] + card[j] + card[k]]

print(max(result))
```
