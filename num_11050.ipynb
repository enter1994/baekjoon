```python
# n과 k가 주어졌을때 nCk를 구하라
```

```python
n, k = map(int, input().split())
```

```python
"""
분자 : n * (n-1) * (n-2) * ... * (n-k+1)
분모 : k * (k-1) * (k-2) * ... * 1
"""
```

```python
mul_x = 1
mul_y = 1

for i in range(k):
    x = n-i
    y = k-i
    mul_x *= x
    mul_y *= y
print(int(mul_x/mul_y))
```
