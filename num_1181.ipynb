n개 단어가 들어옴
1. 길이가 짧은것부터
2. 길이가 같으면 사전 순으로

```python
n = int(input())
```

```python
word = []
for i in range(n):
    word.append(input())
```

```python
word
```

```python
set_word = list(set(word))
```

```python
set_word
```

```python
sort_word = []
```

```python
for i in set_word:
    sort_word.append((len(i), i))
```

```python
sort_word
```

```python
sort_word.sort()
sort_word
```

```python
for i, j in sort_word:
    print(j)
```
